### .dropna     ==> drop None items in df 

```python
df_2 = pd.DataFrame({
    'name': ["ali", "amir", None, "jafar"],
    'age': [15, 16, 20, 44],
    'income(K)': [10, 20, None, 100]
})

df_2
```

```python
df_2.count()
```

```python
df_2.dropna(inplace=True)

df_2
```

```python
df_2 = pd.DataFrame({
    'name': ["ali", "amir", None, "jafar"],
    'age': [15, 16, 20, 44],
    'income(K)': [10, 20, None, 100]
})

df_2
```

```python
df_2.dropna(inplace=True, axis = 'columns')
df_2
```

```python
df_2 = pd.DataFrame({
    'name': ["ali", "amir", None, "jafar"],
    'age': [15, 16, 20, 44],
    'income(K)': [10, 20, None, 100]
})

df_2
```

```python
df_2.dropna(axis = 'index', inplace=True)
df_2
```
